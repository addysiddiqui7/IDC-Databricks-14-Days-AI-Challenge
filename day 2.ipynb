```python
df = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True) #read data into df

print(f"✅ Ready to go! Loaded {df.count():,} events") #modified line to print the number of events
df.show(5)
```

```python
DF = df.select("event_type", "brand", "price") #filtered 3 columns in DF from variable df
DF = DF.filter("price > 100") #filtered out the price in new dataframe DF
DF.show(6) #from now on using DF instead of df to mark the difference after filtering
```

```python
# this groups all rows in DF by the event_type column ("view", "cart", "purchase") and counts the number of events for each type.
DF.groupBy("event_type").count().show()

# the second groupby highlights the most popular brands (by event count) in your filtered data.
top_brands = DF.groupBy("brand").count().orderBy("count", ascending=False).limit(5)

#final count of filtered rows
filtered_count = DF.count()
```
